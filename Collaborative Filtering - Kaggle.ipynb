```python
import pandas as pd
```

```python
movie = pd.read_csv("ml-latest-small/movies.csv")
print(movie.columns)

movie = movie.loc[:,["movieId","title"]]
movie.head(10)
```

```python
rating = pd.read_csv("ml-latest-small/ratings.csv")
rating.columns
```

```python
rating = rating.loc[:,["userId","movieId","rating"]]
rating.head(10)
```

```python
data = pd.merge(movie,rating)
data.head(10)
```

```python
data.shape
```

```python
pivot_table = data.pivot_table(index = ["userId"], columns = ["title"],values = "rating").fillna(data.pivot_table(index = ["userId"], columns = ["title"],values = "rating").mean())
pivot_table.head(15)
```

```python
movie_watched = pivot_table["Bad Boys (1995)"]
print(movie_watched.head())

similarity_with_other_movies = pivot_table.corrwith(movie_watched)  # find correlation between "Bad Boys (1995)" and other movies
similarity_with_other_movies = similarity_with_other_movies.sort_values(ascending=False)
similarity_with_other_movies.head(10)
```
